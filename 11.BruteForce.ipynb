```python
def bf_match(txt:str,subString:str) -> int:
    txtPointer = 0 
    subStringPointer = 0 
    
    while txtPointer != len(txt) and subStringPointer != len(subString):

        if txt[txtPointer] == subString[subStringPointer]: 
            subStringPointer += 1 
            txtPointer += 1 
        else: 
            txtPointer = txtPointer - subStringPointer + 1 
            subStringPointer = 0
    return txtPointer - subStringPointer if subStringPointer == len(subString)  else -1
    

bf_match("abcabcd","abcd")        
```
