```python
from typing import MutableSequence 

def bubbleSort(a:MutableSequence) -> None: 
    n = len(a) - 1 
    
    for i in range(n,0,-1):
        print(f"Iteration {abs(n-i)+1}")
        print(a)
        
        for j in range(i):
            first = a[j]
            second = a[j+1]
            if first>second: 
                a[j],a[j+1] = a[j+1],a[j]
    print("Final")
    print(a)
        
def bubbleSort1(a:MutableSequence) -> None: 
    n = len(a) - 1 
    
    for i in range(n):
        for j in range(n,i,-1):
            first = a[j]
            second = a[j-1]
            
            if first<second: 
                a[j],a[j-1] = a[j-1],a[j]
    print(a)

def bubbleSort2(a:MutableSequence) -> None: 
    n = len(a) - 1 
    exchange = 0 
    for i in range(n,0,-1):
        exchange = 0
        for j in range(i):
            first = a[j]
            second = a[j+1]
            if first>second: 
                a[j],a[j+1] = a[j+1],a[j]
                exchange += 1
        if exchange == 0: 
            break
    return a

            
def bubbleSort3(a:MutableSequence)->None: 
    left = 0 
    right = len(a) - 1 
    last = right 
    while left<right :
        for i in range(right,left,-1):
            if a[i]<a[i - 1]:
                a[i],a[i-1] = a[i-1],a[i]
                last = i 
            left = last 
        
        for i in range(left,right):
            if a[i]>a[i+1]:
                a[i],a[i+1] = a[i+1],a[i]
                last = i 
        right = last
    print(a)

bubbleSort3([5,4,3,2,2,1])
```
