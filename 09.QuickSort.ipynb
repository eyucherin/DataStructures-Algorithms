```python
from typing import MutableSequence 
def partition(a:MutableSequence) -> None: 
    n = len(a)
    left = 0 
    right = n -1 
    x = a[n//2]
    
    while left<right: 
        while a[left]<x:
            left += 1 
        while a[right]>x:
            right -= 1 
        if left <= right: 
            a[left],a[right] = a[right],a[left]
            left += 1 
            right -= 1 
    print(a)   
    print(f"partion left {a[:left]}")
    print(f"partion right {a[left:]}")

partition([1,4,3,5,6,2,3,1,3,2])
```

```python
def qSort(a:MutableSequence, pl: int , pr: int) -> None: 
    left = pl 
    right = pr 
    
    x = a[(left + right)//2]
    
    while left<=right: 
        while a[left]< x:
            left += 1 
        while a[right]>x:
            right -= 1 
        if left <= right: 
            a[left],a[right] = a[right],a[left]
            left += 1 
            right -= 1
    
    if right > pl:
        qSort(a,pl,right)
    if left < pr:
        qSort(a,left,pr)
    print(a)   

qSort([1,4,3,5,6,2,3,1,3,2],0,9)      
```
