```python
def recordEven(myArr):
    p1 = 0 
    p2 = len(myArr) - 1 
    while(p1 < p2):
        if myArr[p1] % 2 == 0: 
            p1 += 1 
        else: 
            myArr[p1],myArr[p2] = myArr[p2],myArr[p1]
            p2 -= 1 
    return myArr
recordEven([1,2,3,4,5,6,7])
```

```python
def dutchPartition(arr,pivot): 
    pivotVal = arr[pivot]
    low,pointer,high = 0, 0, len(arr) - 1
    while (pointer < high):
        if arr[pointer]< pivot: 
            arr[low],arr[pivot] = arr[pivot],arr[low]
            low+= 1
            pointer += 1
        elif arr[pointer] == pivot:
            pointer += 1 
        else: 
            arr[high],arr[pointer] = arr[pointer],arr[high]
            high -= 1 
    return arr 

print(dutchPartition([0,1,2,0,2,1,1],1))
            
```
