```python
from typing import Sequence,Any 
import copy

def BinSearch(a:Sequence,val:int) -> int: 
    low = 0 
    high = len(a) - 1 

    while low <= high: 
        mid = (low+high)//2 
        if a[mid] == val: 
            return mid 
        elif a[mid] < val:
            high = mid - 1 
        elif a[mid] > val: 
            low = mid + 1 
    return -1 


def BinSearch1(a:Sequence,val:int)->int:
    arr = copy.copy(a)
    
    while arr:
        mid = len(arr)//2       
        if arr[mid] == val: 
            return mid 
        elif val>arr[mid]:
            arr[mid+1:]
        elif val<arr[mid]:
            arr[:mid-1]
    return -1

print(BinSearch1([1,2,3,4,5,6,7],-1))

```
