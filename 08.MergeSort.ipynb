```python
def merge(S1,S2,S):
    # i for S1 j for S2
    i = j = 0 

    while i + j < len(S):
        #Copy ith element as next item for S 
        if j == len(S2) or (i<len(S1) and S1[i] < S2[j]):
            S[i+j] = S1[i]
            i+= 1 
        #Copy jth element as next item for S 
        else: 
            S[i+j] = S2[j]
            j+= 1 

def mergeSort(S):
    n = len(S)
    if n < 2: 
        return 
    
    #Divide
    mid = n//2 
    S1 = S[:mid]
    S2 = S[mid:]
    #Conque with recursion 
    mergeSort(S1)
    mergeSort(S2)
    #merge Results 
    merge(S1,S2,S)
    

values = [6,5,4,3,2,2,1]
mergeSort(values)
print(values)
```
