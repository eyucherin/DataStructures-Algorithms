```python
class Node:
    def __init__(self,data):
        self.data = data 
        self.next = None 

class linkedList:
    def __init__(self):
        self.head = None 
    
    def __len__(self):
        count = 0 
        current = self.head
        while current!= None: 
            current = current.next
            count += 1 
        return count
    
    def clear(self):
        self.head = None 
        
    def contains(self,element):
        current = self.head 
        while current: 
            if current.data == element.data:
                return True 
            current  = current.next
        return False 
    
    def getIndex(self,element):
        current = self.head  
        index = 0 
        while current: 
            if current.data == element.data:
                return index 
            index += 1
            current = current.next 
        return -1 
    
    def getFirst(self):
        return self.head.data 
    
    def getLast(self):
        current = self.head 
        while current.next != None: 
            current = current.next 
        return current.data

    def getAt(self,pos): 
        current = self.head 
        count = 0
        while current.next != None and count<pos: 
            current = current.next 
            count += 1 
        return current.data   
        
    def append(self,newVal):
        if self.head == None: 
            self.head = newVal 
            return 
        
        current = self.head 
        while current.next != None: 
            current = current.next
        current.next = newVal 
        
    def insert(self,newVal,position):
        if position == 0: 
            newVal.next = self.head 
            self.head = newVal
            return 
        
        count = 0 
        current = self.head 
        prev = None 
        while count<position and current:
            count += 1 
            prev = current 
            current = current.next 
            
        prev.next = newVal 
        newVal.next = current
    
    def pop(self):
        current = self.head 
        while current.next.next != None: 
            current = current.next 
        current.next = None 
    
    
    def removeElement(self,element):
        found = False 
        current = self.head 
        prev = None 
        
        while current:
            if current.data == element.data: 
                found = True
                break
            prev = current 
            current = current.next 
            
        if prev == None: 
            self.head = current.next 
        elif found:
            prev.next = current.next 
    
    def removeIndex(self,index):
        count = 0 
        current = self.head 
        prev = None
        
        if index > len(self) - 1 or index < 0:
            return
        
        while count < index and current:
            count += 1 
            prev = current
            current = current.next 
        if index == 0: 
            self.head = current.next
        else: 
            prev.next = current.next   
            
    def traverse(self):
        current = self.head 
        answer = []
        while current:
            answer.append(current.data)
            current = current.next 
        return answer 
   
        
    def __str__(self):
        current = self.head
        values = ""
        while current: 
            values += str(current.data) + "->"
            current = current.next 
        return values 

a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)
e = Node(5)
f = Node(6)

myL = linkedList()
myL.append(a)
myL.append(b)
myL.append(c)

myL.insert(d,100)
# print(myL)
# print(myL.getFirst())
# print(myL.getLast())
# print(myL.getAt(-1))
# print(myL.contains(e))
# print(myL.getIndex(e))

print(myL)
#myL.removeElement(a)
myL.removeIndex(-1)
print(myL)
myL.traverse()
```
