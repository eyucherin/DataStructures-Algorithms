```python
def isnthBitPositive(num,position):
    return num & (1<<position)

print(isnthBitPositive(4,2))
```

```python
def setnthBitTo1(num,position):
    num = (num | (1<<position)) 
    return num

print(setnthBitTo1(8,2))
```

```python
def clearNthBit(num,position):
    num = num & ~(1<<position)
    return num 
print(clearNthBit(4,2))
```

```python
def checkOdd(num):
    return (num & 1 == 1)
print(checkOdd(7))
print(checkOdd(8))
```

```python
def checkOppositeSign(a,b):
    return ((a & b) >> 31 == 0)
print(checkOppositeSign(-1,3))
```

```python
def multiplyThreePointFive (a):
    return (a << 1) + a + (a >> 1)

print(multiplyThreePointFive(2))
print(multiplyThreePointFive(5))
```

```python
def turnOffRightMostBit(a):
    return a & (a-1) 

print(turnOffRightMostBit(12))
```

```python
def isPowerOfFour(a):
    return (a & 3 == 0)
print(isPowerOfFour(12))
```

```python
def findMod(n,mod):
    return (n & (mod - 1))
print(findMod(14,3))
```

```python
def find1Element(arr): 
    res = arr[0]
    for i in range(1,len(arr)):
        res = res ^ arr[i]
    print(res)
find1Element([7,3,5,4,5,3,4])
```

```python
def findOddOccuringElement(arr):
    res = arr[0]
    for i in range(1,len(arr)):
        res = arr[i] ^ res 
    return res 

print(findOddOccuringElement([1,2,3,2,3,1,3]))
print(findOddOccuringElement([5,7,2,7,5,2,5]))
```

```python
def count1Bit(n):
    count = 0 
    res = 1 
    while(res < 32):
        val = n & res 
        if val != 0: 
            count += 1 
        res <<= 1 
    return count 

print(count1Bit(6))
# 0110 -> 2
print(count1Bit(8))
# 1000 -> 1
    
```

```python
def countNumOfBitsToFlip(a,b):
    c = a ^ b
    count = 0 
    res = 1
    while(res < 32):
        if(c & res != 0):
            count += 1 
        res <<= 1 
    return count 

print(countNumOfBitsToFlip(10,20))
# 0000 1010, 0001 0100 --> 4 (flip 4 zeros)
print(countNumOfBitsToFlip(7,10))
# 0111, 1010 -> 3 (flip 2 ones and 1 zeros)
```

## The parity of a binary word is 1 if the number of 1s in the word is odd, otherwise it is 0. 

Example1 
word : 1011
parity = 1 

Example2 
word : 1000 1000
parity : 2


```python
def parity1(x):
    result = 0 
    while x: 
        result = result ^ 1 
        x = x & (x-1)
    return result
parity1(2)
```

```python
def parity2 (x):
    count = 0 
    while(x):
        count += x & 1 
        x >>= 1
    return (count % 2)

# 0101 even 
print(parity2(5))
# 1000 odd 
print(parity2(8))
```
