```python
from typing import Any 

class Node: 
    def __init__(self,data):
        self.data = data 
        self.next = None
        self.prev = None 

class DoublyLinkedList: 
    def __init__(self):
        self.head = None 
    
    def __len__(self) -> int:
        current = self.head
        count = 0 
        while current:
            count += 1 
            current = current.next 
        return count 
    
    def append(self,data:int) -> None: 
        node = Node(data)
        if self.head == None: 
            self.head = node 
            return
        current = self.head 
        while current.next: 
            current = current.next
        current.next = node 
        node.prev = current 
    
    def search(self,val:int) -> int: 
        current = self.head 
        indexVal = 0 
        while current: 
            if current.data == val: 
                return indexVal 
            indexVal+= 1 
            current = current.next 
        return -1 
    
    def addFirst(self,val:Any) -> None: 
        node = Node(val)
        current = self.head 
        if current == None: 
            self.head = node 
            return 
        node.next = current 
        current.prev = node 
        self.head = node
    
    def remove(self,val:Any) -> None: 
        current = self.head 
        while current: 
            if current.data == val:
                if current.prev == None: 
                    self.head = current.next 
                else: 
                    current.prev.next = current.next 
                return 
            current = current.next 
        
    
    def __str__(self) -> str:
        current = self.head
        values = ""
        while current:
            values += str(current.data) + "->"
            current = current.next 
        return values 

A = DoublyLinkedList()
A.append(1)
A.append(2)
A.append(3)
A.append(4)
A.append(5)
A.append(6)
A.addFirst(0)
A.remove(6)
print(A)
print(len(A))

B = DoublyLinkedList()
B.addFirst(1)


print(B)

print(A.search(4))
```
