```python
#parent:a[(i - 1) // 2]
#left node: a[i * 2 + 1]
#right node:a[i * 2 + 2] or a[left node + 1 ]

from typing import MutableSequence

def heap_sort(a: MutableSequence) -> None:


    def down_heap(a: MutableSequence, left: int, right: int) -> None:
       
        temp = a[left]      # 루트

        parent = left
        while parent < (right + 1) // 2:
            cl = parent * 2 + 1     # 왼쪽 자식
            cr = cl + 1             # 오른쪽 자식
            child = cr if cr <= right and a[cr] > a[cl] else cl  # 큰 값을 선택합니다.
            if temp >= a[child]:
                break
            a[parent] = a[child]
            parent = child
        a[parent] = temp

    n = len(a)

    for i in range((n - 1) // 2, -1, -1):   # a[i] ~ a[n-1]을 힙으로 만들기
        down_heap(a, i, n - 1)

    for i in range(n - 1, 0, -1):
        a[0], a[i] = a[i], a[0]     # 최댓값인 a[0]과 마지막 원소 a[i]를 교환
        down_heap(a, 0, i - 1)      # a[0] ~ a[i-1]을 힙으로 만들기
    print(a)

heap_sort([6,4,3,7,1,9,8])
```

```python
from typing import MutableSequence 
import heapq, copy 

# O(nlogn)
def heapSort1(a:MutableSequence) -> None: 
    b = copy.copy(a)
    heapq.heapify(b)
    ordered = []
    while b: 
        ordered.append(heapq.heappop(b))
    print(ordered)
    
a = [5,2,7,3,6,1,4]
heapSort1(a)
```

```python
from typing import MutableSequence 
import heapq

# O(logn)
def heapSort2(a:MutableSequence) -> None: 
    heapify = []
    for num in a: 
        heapq.heappush(heapify,num)
    ordered = []
    
    while heapify: 
        ordered.append(heapq.heappop(heapify))
    print(ordered)
heapSort2([5,2,7,3,6,1,4])
```
