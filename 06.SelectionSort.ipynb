```python
from typing import MutableSequence 

def selectionSort(a:MutableSequence) -> None: 
    n = len(a) 
    
    for i in range(n-1):
        minVal = i
        for j in range(i+1,n):
            if a[j]< a[minVal]: 
                minval = j
        a[minval],a[i] = a[i],a[minval]
        
    print(a)
selectionSort([5,4,3,3,2,1])
```
