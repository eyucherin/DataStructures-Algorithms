# EASY
##### Figure out if n'th bit is 0 or 1   val &(1<<pos)!=0

```python
print(4&(1<<2)!=0)

# 4 = 0100
# Figure out it the 3rd bit is 1 or 0 
# shift 1 two to the left 
# --> 0001 
# --> 0100
# 0100 & 0100 = 0100 so True

print(10&(1<<2)!=0)

# 10 = 1010
# Figure out it the 3rd bit is 1 or 0 
# shift 1 two to the left 
# --> 0001 
# --> 0100
# 1010 & 0100 = 0000 so False 
```

##### Set n'th bit to 1   val | (1<<pos)

```python
print(4|(1<<0))
print(10|(1<<2))
```

##### Clear n'th bit  val &~ (1<<pos)

```python
print(4& ~(1<<2))
```

```python
def checkOdd(n):
    return (n&1 == 1)

print(checkOdd(5))
print(checkOdd(12))
```

```python
def checkOppositeSign(x,y):
    return ((x^y) >> 31 != 0 )

# if True sign is opposite, otherwise sign in the same 
print(checkOppositeSign(8,-8))
```

```python
def addOne(x):
    a = 1
    while(a & x):
        x = x ^ a
        a <<= 1 
    x = x ^ a
    print(x)
    
addOne(13)
```

```python
def multiplyThreePointFive(x):
    return (x<<1) + x + (x>>1)

print(multiplyThreePointFive(2))
print(multiplyThreePointFive(5))
```

```python
#MySOL
def turnOffRightMostBit1(x):
    count = 1 
    while(count & x == 0):
        count <<= 1 
    return (x ^ count)
print(turnOffRightMostBit1(12))
# 0000 1100
# 0000 1000

def turnOffRightMostBit2(x):
    return x & (x-1)
print(turnOffRightMostBit2(12))
```

```python
def isPowerOfFour(x):
    return (x&3 == 0)
print(isPowerOfFour(64))
print(isPowerOfFour(75))
```
