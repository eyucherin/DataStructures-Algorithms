```python
def factorial(n):
    if n == 1: 
        return 1 
    else: 
        return n * factorial(n-1)

if __name__ == "__main__":
    print(factorial(3))
```

```python
import math 
def GCD(x,y):
    if y == 0:
        return x
    else: 
        return GCD(y,x%y)
    
print(GCD(30,108))
print(GCD(108,30))
print(math.gcd(30,108))
print(math.gcd(108,30))
```

```python
def towerOfHanaoi(number:int,x:int,y:int):
    if number> 1:
        towerOfHanaoi(number-1,x,6-x-y)
        
    print(f"{number} was moved from {x} to {y}")
    
    if number>1:
        towerOfHanaoi(number-1,6-x-y,y)
towerOfHanaoi(3,1,3)
print()
towerOfHanaoi(4,1,3)
```

```python
def fib(n):
    if n == 0:
        return 0 
    elif n == 1:
        return 1 
    else:
        return fib(n-1) + fib(n-2)
print(fib(5))
```
