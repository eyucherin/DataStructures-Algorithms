```python
from collections import deque

deq = deque([1,2,3,4,5])
deq.append(6)
deq.appendleft(0)
deq.extendleft([-1,-2,-3])
print(deq.popleft())
print(deq)
print(list(deq))


deq1 = deque([1,2,3,4,5,6,7,8]) 
deq1.rotate(3)
print(deq1)# rotate to the right
deq1.rotate(-3)  
print(deq1)# rotate to the left 
```
